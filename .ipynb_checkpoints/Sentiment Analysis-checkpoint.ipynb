#### NLTK

```python
#importing the libraries
import pandas as pd
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import nltk
nltk.download('vader_lexicon')

# corpus 
texts = ["I love this product, it's amazing!",
         "This is the worst experience I've ever had",
         "It's okay, not great but not bad either", 
         "I am extremely happy with the service!",
         "I hate it when things don't work as expected", 
         "This product exceeded my expectations! The quality is outstanding and it's very easy to use.",
         "I'm not very happy with this product. It stopped working after just two weeks.",
         "Great value for the price. Works perfectly and I would definitely recommend it to others.",
         "The product is okay, but the customer service was terrible. They were not helpful at all.",
         "I've been using this for a month now and it's fantastic. No issues and very reliable.",
         "Not worth the money. It's poorly made and feels very cheap.",
         "Excellent product! Arrived on time and in perfect condition. Very satisfied.",
         "The product does what it says, but I was expecting better performance.",
         "I love this product! It has made my life so much easier. Five stars!",
         "Decent product, but there are better options available at a similar price point."]

# creating a dataframe
df = pd.DataFrame(texts, columns=["text"])

# declaring the sentiment analyzer
sia = SentimentIntensityAnalyzer()

#function to get the sentiment
def get_sentiment(text):
    return sia.polarity_scores(text)

#applying the function to the dataframe
df["sentiment"] = df["text"].apply(get_sentiment)

# splitting the sentiment into 4 columns
df = pd.concat([df.drop(['sentiment'], axis=1), df['sentiment'].apply(pd.Series)], axis=1)

def label(row):
    scores = [row['neg'], row['neu'], row['pos']]
    score = max(scores)
    if score == row['neg']:
        return 'negative'
    elif score == row['neu']:
        return 'neutral'
    else:
        return 'positive'
    
df['label'] = df.apply(lambda row: label(row), axis=1) 
#displaying the results
df
```

#### TextBlob

```python
from textblob import TextBlob
import pandas as pd 

# corpus
texts = ["I love this product, it's amazing!",
         "This is the worst experience I've ever had",
         "It's okay, not great but not bad either", 
         "I am extremely happy with the service!",
         "I hate it when things don't work as expected", 
         "This product exceeded my expectations! The quality is outstanding and it's very easy to use.",
         "I'm not very happy with this product. It stopped working after just two weeks.",
         "Great value for the price. Works perfectly and I would definitely recommend it to others.",
         "The product is okay, but the customer service was terrible. They were not helpful at all.",
         "I've been using this for a month now and it's fantastic. No issues and very reliable.",
         "Not worth the money. It's poorly made and feels very cheap.",
         "Excellent product! Arrived on time and in perfect condition. Very satisfied.",
         "The product does what it says, but I was expecting better performance.",
         "I love this product! It has made my life so much easier. Five stars!",
         "Decent product, but there are better options available at a similar price point."]

# creating a dataframe
df = pd.DataFrame(texts, columns=["text"])

# function to get the sentiment
def get_sentiment(text):
    blob =  TextBlob(text)
    return blob.sentiment.polarity, blob.sentiment.subjectivity

# applying the function to the dataframe
df[["Polarity", "Subjectivity"]] = df["text"].apply(get_sentiment).apply(pd.Series)

# function to label the sentiment
df['Label'] = df['Polarity'].apply(lambda x: 'Positive' if x > 0 else ('Negative' if x < 0 else 'Neutral'))

# displaying the results
df
```

#### Afinn

```python
from afinn import Afinn
import pandas as pd 

# corpus
texts = ["I love this product, it's amazing!",
         "This is the worst experience I've ever had",
         "It's okay, not great but not bad either", 
         "I am extremely happy with the service!",
         "I hate it when things don't work as expected", 
         "This product exceeded my expectations! The quality is outstanding and it's very easy to use.",
         "I'm not very happy with this product. It stopped working after just two weeks.",
         "Great value for the price. Works perfectly and I would definitely recommend it to others.",
         "The product is okay, but the customer service was terrible. They were not helpful at all.",
         "I've been using this for a month now and it's fantastic. No issues and very reliable.",
         "Not worth the money. It's poorly made and feels very cheap.",
         "Excellent product! Arrived on time and in perfect condition. Very satisfied.",
         "The product does what it says, but I was expecting better performance.",
         "I love this product! It has made my life so much easier. Five stars!",
         "Decent product, but there are better options available at a similar price point."]

# creating a dataframe
df = pd.DataFrame(texts, columns=["text"])

# creating the object for the AFINN lexicon
afinn = Afinn()

# function to get the sentiment
def get_sentiment(text):
    return afinn.score(text)

# applying the function to the dataframe
df["Score"] = df["text"].apply(get_sentiment).apply(pd.Series)

# function to label the sentiment
df['Label'] = df['Score'].apply(lambda x: 'Positive' if x > 0 else ('Negative' if x < 0 else 'Neutral'))

# displaying the results
df
```

```python
# importing the libraries
import pandas as pd
from textblob import TextBlob
from matplotlib import pyplot as plt


# importing the dataset
data = pd.read_csv(r"C:\Data Storage\Natural Language Processing (NLP)\Data\Tweets.csv")

# getting the tweets
tweets = data[['text']].copy()

# cleaning the tweets
tweets['text'] = tweets['text'].str.strip()
tweets['text'] = tweets['text'].astype('str').fillna('')

#function to get the sentiment
def get_sentiment(text):
    blob =  TextBlob(text)
    return blob.sentiment.polarity, blob.sentiment.subjectivity


tweets["sentiment"] = tweets['text'].apply(get_sentiment)

# applying the function to the dataframe
tweets[["Polarity", "Subjectivity"]] = tweets["text"].apply(get_sentiment).apply(pd.Series)

# function to label the sentiment
tweets['Label'] = tweets['Polarity'].apply(lambda x: 'Positive' if x > 0 else ('Negative' if x < 0 else 'Neutral'))

tweets['Emotions'] = tweets['Subjectivity'].apply(lambda x: 'Objective' if x < 0.5 else 'Subjective')

# plotting the sentiment distribution
tweets['Label'].value_counts().plot(kind='bar')
plt.title('Sentiment Distribution on Bar Graph')
plt.show()

plt.pie(tweets['Label'].value_counts(), labels=tweets['Label'].value_counts().index, autopct='%1.1f%%')
plt.title('Pie Chart of Sentiment Distribution')
plt.show()
```

```python
corpus = ""
for text in tweets[(tweets['Emotions'] == 'Objective') & ( tweets['Label'] == 'Positive')]['text']:
    corpus += text + " "
```

```python
import nltk 
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from collections import Counter
# tokenizing the corpus
words = word_tokenize(corpus)

# getting the list of stop words
stop_words = set(stopwords.words('english'))

# removing the stop words and puntuation marks from the corpus
# words = [word for word in words if word.isalnum() and word not in stop_words]

words = [word for word in words if word not in stop_words]
words = [word for word in words if word.isalnum()]
words = [word for word in words if word != "I"]

words = [word.lower() for word in words]
#calculate word frequency
words_frequency = Counter(words)

#print the most common words
print('Word Frequency:')
for word , freq in words_frequency.most_common(20):
    print(f"{word}: {freq}")
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

#Generate and display a word cloud
wordCloud = WordCloud(width = 800 , height = 400 , background_color='white').generate_from_frequencies(words_frequency)

#Plot the word cloud
plt.figure(figsize=(10,5))
plt.imshow(wordCloud, interpolation='bilinear')
plt.title('Word Cloud')
plt.axis('off')
plt.show()
```

##### NLTK

```python
# importing the libraries
import pandas as pd
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import nltk
from matplotlib import pyplot as plt
nltk.download('vader_lexicon')

# importing the dataset
data = pd.read_csv(r"C:\Data Storage\Natural Language Processing (NLP)\Data\Tweets.csv")

# getting the tweets
tweets = data[['text']].copy()

# cleaning the tweets
tweets['text'] = tweets['text'].str.strip()
tweets['text'] = tweets['text'].astype('str').fillna('')

# declaring the sentiment analyzer
sia = SentimentIntensityAnalyzer()

#function to get the sentiment
def get_sentiment(text):
    return sia.polarity_scores(text)

tweets["sentiment"] = tweets['text'].apply(get_sentiment)

# splitting the sentinment into 4 columns
tweets = pd.concat([tweets.drop(['sentiment'], axis=1), tweets['sentiment'].apply(pd.Series)], axis=1)

# function for labeling the sentiment
def label(row):
    scores = [row['neg'], row['neu'], row['pos']]
    score = max(scores)
    if score == row['neg']:
        return 'negative'
    elif score == row['neu']:
        return 'neutral'
    else:
        return 'positive'
   
# applying the function of labelling to the dataframe 
tweets['label'] = tweets.apply(label, axis=1) 

# plotting the sentiment distribution
tweets['label'].value_counts().plot(kind='bar')
plt.title('Sentiment Distribution on Bar Graph')
plt.show()

plt.pie(tweets['label'].value_counts(), labels=tweets['label'].value_counts().index, autopct='%1.1f%%')
plt.title('Pie Chart of Sentiment Distribution')
plt.show()
```
